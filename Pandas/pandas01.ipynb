# Pandas Series
A Series is a one-dimensional labeled array capable of holding any data type. the axis labels are collectively called the index.

# Creating Series
there are multiple ways to create a pandas series

```python
import numpy as np
import pandas as pd
```

```python
labls = ['a','b','c']
lists = [10,20,30]
arr = np.array([17,25,34])
d = {'e':10,'f':20,'g':30}
```

```python
pd.Series(lists)   # (Series --> veritcal)
```

```python
# Custom labels
pd.Series(lists,index = labls)
```

```python
# Series using a array
pd.Series(arr)
```

```python
# Series using the dictionary
pd.Series(d)  #no need to pass the index already key took the place of index
```
