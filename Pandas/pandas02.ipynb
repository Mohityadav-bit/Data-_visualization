# DataFrame 


Creating a DataFrame

```python
import numpy as np
import pandas as pd
```

```python
# DataFrame using the dictionary
data = {
    'Name': ['John','Anna','Peter','Linda'],
    'Age' :[28,34,29,42],
    'City':['New York','Paris','Berlin','London'],
    'Salary':[6500,7000,6200,8500]    
}
df = pd.DataFrame(data)
df
```

```python
# DataFrame using the list
data_list = [
    ['Jhon',28,'New York',65000],
    ['Anna',34,'Paris',70000],
    ['Peter',29,'Berlin',62000],
    ['Linda',42,'London',85000]
]
df1 = pd.DataFrame(data_list)
cols = ['Name','Age','City','Salary']
# Adding the col name
df1 = pd.DataFrame(data_list,columns=cols)
```

```python
df1
```

```python
df1['Occupation'] = ['Farmer','Doctor','Engineer','Doctor']
df1
```

Removing Columns

```python
df1.drop('Occupation',axis=1,inplace=True)  #inplace= Ture removes the data competely form the memory as well
```

```python
# Remoing multiple data
df1.drop(['City','Age'],axis=1)  #axis=0 for horizontal ||| axis=1 for cols
```
