```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x = np.linspace(0,5,11)
x
# Broadcasting
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Sample Data")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

Subplots in Matplotlib

```python
plt.subplot(2,2,1)   
plt.plot(x,x)
plt.subplot(2,2,4)  
plt.plot(y,x**3)
plt.subplot(2,2,2)   
plt.plot(x,y)
plt.subplot(2,2,3)  
plt.plot(y,x) 
```

Object Oriented Way..

Matplotlib allow thw aspect ratio, DPI and figure size to be specified when the figure object is created. u can use the figsize and dpi keywords arguments.

figsize is a tuple of the width and height of the figure in the inches dpi is the dots-per-inches (pixels per inches)

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis1 = fig.add_axes([0.1,0.1,0.4,0.3])
axis1.plot(x,y)
```

Types of plots

```python
# scatter plots
plt.scatter(x,y)
```

```python
# Histogram
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
# Box plot

plt.boxplot(x)
```

 Saving plots


```python
fig = plt.figure()
axe = fig.add_axes([0.1,0.1,0.8,0.8])
plt.title("Basic Plot")
axe.plot(x,y)
plt.savefig("basicplot.png")
```
