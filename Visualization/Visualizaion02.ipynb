```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Distribution Plots:-- we use distribution plots to visualize the distribution of quantitative data.

plots that allow us to visulaize the distribution of a data set. these plots are:-

--> distplot/histplot
--> jointplot
--> pairplot
--> rugplot



```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

qualitative and quatitaive

Distribution plots

```python
sns.histplot(df['total_bill'],bins=20)
```

```python
# Karnal data estimater
sns.histplot(df['total_bill'],bins=20,kde = True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde = True)
```

Joint plots

```python
sns.jointplot(x= 'total_bill', y = 'tip', data = df, kind= 'scatter')
```

```python
sns.jointplot(x= 'total_bill', y = 'tip', data = df, kind= 'kde')
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.rugplot(df)
```
