# deep and shallow copy
| Feature       | Shallow Copy                    | Deep Copy                      |
| ------------- | ------------------------------- | ------------------------------ |
| Definition    | Copies only outer structure     | Copies everything recursively  |
| Memory        | Shares inner objects            | Creates completely new objects |
| Change Effect | Changes may reflect in original | No effect on original          |
| Speed         | Faster                          | Slower                         |
| Use Case      | When data is simple             | When data is nested            |

Stacking Array

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
```

```python
a
```

```python
b
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```
