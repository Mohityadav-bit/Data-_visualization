## NumPy Random Functions Comparison

| Function | Description | Parameters | Range of Values | Distribution | Example |
|----------|------------|------------|----------------|--------------|---------|
| rand() | Generates random numbers | rand(d0, d1, ..., dn) → shape as separate arguments | [0, 1) | Uniform distribution | np.random.rand(2,3) |
| randn() | Generates random numbers | randn(d0, d1, ..., dn) → shape as separate arguments | (-∞, +∞) | Normal (Gaussian) distribution (mean=0, std=1) | np.random.randn(2,3) |
| randint() | Generates random integers | randint(low, high, size) → range + shape | [low, high) | Uniform (discrete) | np.random.randint(1, 10, (2,3)) |

Array Methods
Array methods in NumPy are functions applied on arrays to perform operations like reshaping, aggregation, sorting, and type conversion.



## NumPy Array Methods

|--------|------------|---------|

| arr.max() | Maximum value | arr.max() |
| arr.min() | Minimum value | arr.min() |
| arr.argmax() | Index of max value | arr.argmax() |

## Difference Between Array Attributes and Array Methods

| Feature | Array Attributes | Array Methods |
|--------|-----------------|---------------|
| Definition | Properties that describe the array | Functions that perform operations on the array |
| Usage | Accessed without parentheses | Called using parentheses |
| Syntax | arr.shape | arr.reshape() |
| Purpose | Provide information about array | Modify or compute values |
| Example | arr.ndim, arr.dtype | arr.sum(), arr.sort() |
| Output | Returns metadata (info) | Returns processed data or modifies array |

Numpy Indexing and Slicing of vectors


```python
narr = np.arange(11,21)
narr
```

```python
narr[0]   # here we can directly retrive the element by the index of the element same indexing as he list
```

```python
narr[1:5]
```

```python
# Assume example:
# narr = np.array([10, 20, 30, 40, 50, 60])

narr[:]      # Selects ALL elements (full copy/view of array)

narr[1:]     # Selects elements from index 1 to end
             # → [20, 30, 40, 50, 60]

narr[3::2]   # Starts from index 3, goes till end, step = 2
             # → Picks every 2nd element from index 3
             # → [40, 60]
narr[:3]     # first 3 elements → [10, 20, 30]
narr[::-1]   # reverse array → [60, 50, 40, 30, 20, 10]
```
