🔹 What is NumPy (in simple terms)?

NumPy = Numerical Python

It’s a core Python library used for:

Fast mathematical operations
Handling large datasets
Working with arrays (like a more powerful version of lists)
🔹 Why NumPy was developed (the real problem it solved)

Before NumPy, Python had a big limitation:

👉 Python lists were slow and inefficient for numerical work

Example problem:

Lists store mixed data types (int, float, string)
Operations like addition/multiplication require loops → very slow
❌ Without NumPy:
arr = [1, 2, 3]
result = []

for i in arr:
    result.append(i * 2)
✅ With NumPy:
import numpy as np
arr = np.array([1, 2, 3])
result = arr * 2

👉 NumPy uses optimized C code internally, making it MUCH faster.

🔹 History of NumPy (short + important)
1990s: Python used for general programming, not scientific computing
Numeric library was created for arrays
Later Numarray was developed for better performance
Problem: Two libraries → confusion 😵

👉 Then came:

Travis Oliphant (creator of NumPy)
In 2005, he combined Numeric + Numarray → NumPy was born

👉 Today:

NumPy is the foundation of the entire data science ecosystem
Libraries like:
Pandas
Matplotlib
Scikit-learn
all depend on NumPy

## Python List vs NumPy Array

| Feature | Python List | NumPy Array |
|--------|------------|-------------|
| Library | Built-in | Requires NumPy |
| Data Type | Mixed types allowed | Homogeneous (same type) |
| Performance | Slower | Faster (C optimized) |
| Memory Usage | More memory | Less memory |
| Operations | Requires loops | Vectorized operations |
| Multidimensional | Nested lists | Built-in support |
| Math Functions | Limited | Rich functions (mean, sum, etc.) |
| Use Case | General purpose | Data science, ML |
